```python
import time
import requests
from bs4 import BeautifulSoup    
```

```python
def price_tracker(stockcode):
    
    stockvalue = ""
    url = "https://in.finance.yahoo.com/quote/%5ENSEBANK?p=^" + stockcode
    print(url)
    response= requests.get(url)
    soup = BeautifulSoup (response.text,'html.parser')
    
    if len(soup.find_all('span', {'class' : 'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)' })) > 0:
        stockvalue = soup.find_all('span', {'class' : 'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)' })[0].text
        return (stockvalue)
    else:
        stockvalue = soup.find_all('span', {'class' : 'Trsdu(0.3s) Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(b)' })[0].text
        return (stockvalue)
```

```python
while True:
    print("current price : ", price_tracker('NSEBANK'))
    time.sleep(5)
```
